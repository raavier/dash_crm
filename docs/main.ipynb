```python
# Initialize Databricks Connect session with profile and cluster
from databricks.connect import DatabricksSession
from databricks.sdk.core import Config
from IPython.core.magic import register_line_cell_magic
from IPython.display import display

# Configure using the 'ravi-local' profile and cluster ID
config = Config(
    profile="ravi-local",
    cluster_id="0103-144058-4tvp4kpg"
)

# Create Spark session
spark = DatabricksSession.builder.sdkConfig(config).getOrCreate()

# Register %sql magic command
@register_line_cell_magic
def sql(line, cell=None):
    """Execute SQL query using Spark"""
    query = cell if cell else line
    df = spark.sql(query)
    display(df.toPandas())
    return df

print("✓ Spark session initialized successfully!")
print(f"✓ Spark version: {spark.version}")
print(f"✓ Connected to: {config.host}")
print(f"✓ Using cluster: {config.cluster_id}")
print("✓ %sql magic command registered!")
```

```python
# Carregar metadados das tabelas CRM
df = spark.sql("""SELECT * FROM hs_franquia.gold_connect_bot.table_metadata""")

# Exibir os dados
display(df)

# Salvar em CSV para documentação
df.toPandas().to_csv('crm_table_metadata.csv', index=False)
print(f"\n✓ Metadados salvos em 'crm_table_metadata.csv'")
print(f"✓ Total de registros: {df.count()}")
```
